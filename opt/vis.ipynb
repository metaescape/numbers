```python
%reload_ext autoreload
%autoreload 2
import seaborn as sns
import matplotlib.pyplot as plt

# read data from data/W_true.csv
import numpy as np
W_true = np.loadtxt('data/W_true.csv', delimiter=',')
```

```python
from notears import solve

W = solve()
```

```python
# create 2x1 subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# plot W_true and W
sns.heatmap(W_true, cmap='coolwarm', center=0, xticklabels=False, yticklabels=False, ax=axs[0])
axs[0].set_title('W_true')

sns.heatmap(W, cmap='coolwarm', center=0, xticklabels=False, yticklabels=False, ax=axs[1])
axs[1].set_title('W')
    
```

```python
# plot 1x3 subplots

thresholds = [0, 0.1, 0.3, 1]
n = len(thresholds)
fig, axs = plt.subplots(1, n, figsize=(n*5, 4))


for j, threshold in enumerate(thresholds):
    # binary W
    W_bin = (W > threshold).astype(int)
    # heatmap

    sns.heatmap(W_bin, cmap='coolwarm', center=0, xticklabels=False, yticklabels=False, ax=axs[j])
    axs[j].set_title('Threshold = {}'.format(threshold))
```
