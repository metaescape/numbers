```python
# read data from data/nongaussian_samples.csv, X1 for first column, X2 for second column


import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('data/nongaussian_samples.csv', delimiter=' ', skip_header=0)
print(data.shape)
```

```python
X1 = data[:, 0]
X2 = data[:, 1]

# plot the data
plt.scatter(X1, X2)
```

```python
# regressing X2 on X1 without intercept
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)
model.fit(X1[:, np.newaxis], X2)
print(model.coef_)

# plot (X1, X2 − bX1)
plt.scatter(X1, X2 - model.coef_[0] * X1)
```

```python
model = LinearRegression(fit_intercept=False)
model.fit(X2[:, np.newaxis], X1)
print(model.coef_)

# plot (X2, X1 − bX2)
plt.scatter(X2, X1 - model.coef_[0] * X2)
```

You observe data from either the SCM Ma:
X1 = ε1

X2 = βX1 + ε2

where ε1⊥ε2

or the SCM Mb:

X2 = ε2

X1 = βX2 + ε1

where ε1⊥ε2

以上两个模型，哪个是正确的？

注意 X1 和 X2-bX1 残差是独立的，这相当于 X1 和 e2 是独立的，符合 Ma 假设，


而 X2 和 X1-bX1 的残差却不是独立的,因为在不同 X1 取值下残差取值范围不同（但根据线性回归定义X2 和 X1-bX1 是不相关的）,不满足 Mb 假设。

因此 Ma 更符合数据的生成过程。
