```python
# read data from data/nongaussian_samples.csv, X1 for first column, X2 for second column


import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('data/nongaussian_samples.csv', delimiter=' ', skip_header=0)
print(data.shape)
```

```python
X1 = data[:, 0]
X2 = data[:, 1]

# plot the data
plt.scatter(X1, X2)
```

```python
# regressing X2 on X1 without intercept
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)
model.fit(X1[:, np.newaxis], X2)
print(model.coef_)

# plot (X1, X2 âˆ’ bX1)
plt.scatter(X1, X2 - model.coef_[0] * X1)
```

```python
model = LinearRegression(fit_intercept=False)
model.fit(X2[:, np.newaxis], X1)
print(model.coef_)

# plot (X2, X1 âˆ’ bX2)
plt.scatter(X2, X1 - model.coef_[0] * X2)
```
